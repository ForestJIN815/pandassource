```python
values = [30,25,20,13,10,2]
labels = ["python","java","script", "c#", "c++", "etc"]

# startangle : 원하는 각도에서 시작할 수 있으나 각도를 주면 시계 반대 방향으로 그려짐
# counterclock=False : 시계 방향으로 원 작성
plt.pie(values, labels=labels, startangle=90,  counterclock=False, autopct="%.1f%%");
```

```python
def custom_autopct(pct):
    return ("%.1f%%" % pct) if pct >=10 else ""

wedgeprops = {"width":0.8, "edgecolor":"w", "linewidth":3}
plt.pie(values, labels=labels, startangle=90,  counterclock=False, autopct=custom_autopct, colors=colors, wedgeprops=wedgeprops);
```

```python
# 데이터프레임 생성

data = {
    "영화":["명량","극한직업","신과함께 - 죄와 벌","국제시장","베테랑","괴물","도둑들","7번방의 선물","암살","광해 - 왕이 된 남자"],
    "개봉연도":[2014,2019,2017,2014,2015,2006,2012,2013,2015,2012],
    "관객수":[1761,1626,1441,1425,1341,1301,1298,1281,1270,1231],
    "평점":[8.88,9.20,8.73,9.16,9.24,8.62,7.64,8.83,9.10,9.25]
}

movie_df= pd.DataFrame(data)
movie_df
```

```python
# 영화 데이터프레임을 이용한 차트 작성
# x축은 영화명, y축은 평점인 bar

plt.bar(movie_df["영화"], movie_df["평점"]);
```

```python
# 차트 세부설정
# 제목 : 국내 Top 10 영화 평점 정보
# x축 레이블명 : 영화
# y축 레이블명 : 평점
# x축 영화명은 45도 각도로 설정

plt.bar(movie_df["영화"], movie_df["평점"])
plt.title("국내 Top 10 영화 평점 정보")
plt.xlabel("영화")
plt.ylabel("평점")
plt.xticks(rotation=45);
```

```python
# 개봉 연도별 평점 변화 추이를 선 차트로 작성
# 연도별 평균 평점을 구한 뒤 그래프로 작성
# x축: 개봉연도, y축: 평점

# 연도별 평균 평점 구하기
# movie_df.groupby("개봉연도")["평점"].mean() 시리즈구조

df1 = movie_df.groupby("개봉연도")["평점"].mean().reset_index()
df1

# 그래프 작성
plt.plot(df1["개봉연도"], df1["평점"]);
```

```python
plt.plot(df1["개봉연도"], df1["평점"], marker="o");
plt.xticks([2005,2010,2015,2020])
plt.ylim(top=10, bottom=7);
```

```python
# 평점이 9점 이상인 영화의 비율을 확인할 수 있는 원 차트 작성

# 9점 이상
condition = movie_df["평점"] >= 9.0
condition
```

```python
# 9점 이상
movie_df[condition]

# 9점 이하
movie_df[~condition]
```
